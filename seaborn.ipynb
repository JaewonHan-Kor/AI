```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
sns.__version__
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
df.describe(include='all')
```

```python
#total_bill 질량 분포
sns.histplot(x=df['total_bill'])
plt.show()
```

```python
#total_bill과 tip에 관한 질량 분포
sns.histplot(x=df['total_bill'],y=df['tip'])
plt.show()
```

```python
#단일 변수(total_bill)에 대한 질량 밀도(추정) 그래프
sns.kdeplot(x=df['total_bill'])
plt.show()
```

```python
#두개의 변수에 대한 질량 밀도(추정) 그래프
sns.kdeplot(x=df['total_bill'],y=df['tip'])
plt.show()
```

```python
#단일 변수에 대한 ecdf plot : 누적 분포를 시각화
sns.ecdfplot(x=df['total_bill'])
plt.show()
```

```python
sns.kdeplot(x=df['total_bill'])
sns.rugplot(x=df['total_bill'])
plt.show()
```

```python
#Bar Plot : 이변량(bivariate) 분석을 위한 plot
sns.barplot(x=df['sex'],y=df['tip'])
plt.show()
```

```python
#Count Plot : 범주형 변수의 발생 횟수
sns.countplot(x=df['sex'])
plt.show()
```

```python
#Box Plot : 최대(maximum), 최소(minimum), 평균(mean), 1 사분위수(first quartile), 3 사분위수(third quartile)
#특이치(outliar)를 발견하기에도 좋다
sns.boxplot(x=df['total_bill'])
plt.show()
```

```python
sns.boxplot(y=df['total_bill'],x=df['smoker'])
plt.show()
```

```python
#Violin Plot: Box Plot과 비슷하지만 분포에 대한 보충 정보가 제공
sns.violinplot(y=df['total_bill'],x=df['smoker'])
plt.show()
```

```python
#Strip Plot : 연속형 변수와 범주형 변수 사이의 그래프
sns.stripplot(y=df['total_bill'],x=df['smoker'])
plt.show()
```

```python
#Swarm Plot : Strip plot과 violin plot의 조합
#데이터 포인트 수와 함께 각 데이터의 분포도 제공
sns.swarmplot(y=df['total_bill'],x=df['smoker'])
plt.show()
```

```python
#Heat Map :데이터 간의 수치에 따라 색상을 입힘으로써 직관적인 통찰을 얻을 수 있다
sns.heatmap(df.corr(),annot=True, cmap='viridis')
plt.show()
```

```python
#Cluster Map : 행렬 데이터를 가지고 있고, 유사성에 따라 몇몇 특징들을 그룹화하기 원한다면 cluster map 사용
sns.clustermap(df.corr(), annot=True, cmap='viridis')
plt.show()
```

```python
#Facet Grid : 어떠한 조건에 따라 그래프를 각각 확인해보고 싶을 때 사용
sns.FacetGrid(df, col = 'smoker',row='sex').map(sns.distplot,'total_bill')
plt.show()
```

```python
#Joint Plot : 두 변수에 대한 displot의 조합
sns.jointplot(x=df['total_bill'],y=df['tip'],kind='scatter')
plt.show()
```

```python
#Pair Plot : 데이터셋을 통째로 넣으면 숫자형 특성에 대하여 각각에 대한 히스토그램과
# 두 변수 사이의 scatter plot을 그린다
sns.pairplot(df)
```

```python
#Reg Plot : Regression 결과를 그래프로 보여준다
sns.regplot(x='tip', y='total_bill', data=df)
plt.show()
```

```python
#LM Plot : regplot()과 faceGrid를 결합
sns.lmplot(x='tip', y='total_bill', data=df, hue='sex')
plt.show()
```
